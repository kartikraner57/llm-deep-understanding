```python
import re

# 1️⃣ Text and Vocabulary
text = [ 
    'All that we are is the result of what we have thought',
    'To be or not to be that is the question',
    'Be yourself everyone else is already taken'
]

# Combine all words & lowercase
allwords = re.split(r'\s', ' '.join(text).lower())

# Unique sorted vocab
vocab = sorted(set(allwords))

# Encoder & Decoder dictionaries
word2idx = {word:i for i, word in enumerate(vocab)}
idx2word = {i:word for i, word in enumerate(vocab)}

# 2️⃣ Encoder function
def encoder(sentence):
    words = re.split(r'\s', sentence.lower())
    return [word2idx[word] for word in words]

# 3️⃣ Decoder function
def decoder(indices):
    return ' '.join([idx2word[i] for i in indices])

# 4️⃣ Test sentence
newtext = 'we already are the result of what everyone else already thought'

# Encode
newtext_tokenIDs = encoder(newtext)

# Decode
decoded_text = decoder(newtext_tokenIDs)

# 5️⃣ Print results
print('Original text:')
print(f'\t{newtext}')

print(f'\nToken IDs:')
print(f'\t{newtext_tokenIDs}')

print(f'\nDecoded text:')
print(f'\t{decoded_text}')
```

```python
import numpy as np
import re

# Text and vocab setup (from previous exercises)
text = [ 
    'All that we are is the result of what we have thought',
    'To be or not to be that is the question',
    'Be yourself everyone else is already taken'
]

allwords = re.split(r'\s', ' '.join(text).lower())
vocab = sorted(set(allwords))
word2idx = {word:i for i, word in enumerate(vocab)}
idx2word = {i:word for i, word in enumerate(vocab)}

def encoder(sentence):
    words = re.split(r'\s', sentence.lower())
    return [word2idx[word] for word in words]

# Tokens of all text
alltext = ' '.join(text)
tokens = encoder(alltext)

# Target word
targetWord = 'to'
targetIdx = word2idx[targetWord]

# Find locations
targetLocs = np.where(np.array(tokens) == targetIdx)[0]
print(f'"{targetWord}" appears at indices {targetLocs}\n\n')

# Print context
for t in targetLocs:
    token_window = tokens[t-1:t+2]  # previous, current, next
    word_window = [idx2word[i] for i in token_window]
    print(token_window, word_window)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import re

# Text & vocab
text = [ 
    'All that we are is the result of what we have thought',
    'To be or not to be that is the question',
    'Be yourself everyone else is already taken'
]

allwords = re.split(r'\s', ' '.join(text).lower())
vocab = sorted(set(allwords))
word2idx = {word:i for i, word in enumerate(vocab)}

# 1️⃣ One-hot matrix
word_matrix = np.zeros((len(allwords), len(vocab)), dtype=int)
for i, word in enumerate(allwords):
    word_matrix[i, word2idx[word]] = 1

# 2️⃣ Show the matrix
print(f'One-hot encoding matrix is of size {word_matrix.shape}\n')
print(word_matrix)

# 3️⃣ Visualize
_, ax = plt.subplots(1, figsize=(10,5))
plt.imshow(word_matrix, cmap='gray', origin='lower', aspect='auto')
ax.set(xlabel='Word index', yticks=range(len(vocab)))
ax.grid(linestyle='--', axis='y')
plt.show()
```
